```python
import pickle
import os

file_name_BL = 'baseline_CNN.pkl' 
file_name_no_BN_MP = 'baseline_CNN_no_batchnorm_maxpool.pkl' 
file_name_no_BN = 'baseline_CNN_no_batchnorm.pkl' 
file_name_no_MP = 'baseline_CNN_no_maxpool.pkl' 

with open(os.getcwd() + '/../results/' + file_name_BL, 'rb') as file:
    tr_acc_BL = pickle.load(file)
    te_acc_BL = pickle.load(file)
    tr_loss_BL = pickle.load(file)
    te_loss_BL = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_no_BN_MP, 'rb') as file:
    tr_acc_no_BN_MP = pickle.load(file)
    te_acc_no_BN_MP = pickle.load(file)
    tr_loss_no_BN_MP = pickle.load(file)
    te_loss_no_BN_MP = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_no_BN, 'rb') as file:
    tr_acc_no_BN = pickle.load(file)
    te_acc_no_BN = pickle.load(file)
    tr_loss_no_BN = pickle.load(file)
    te_loss_no_BN = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_no_MP, 'rb') as file:
    tr_acc_no_MP = pickle.load(file)
    te_acc_no_MP = pickle.load(file)
    tr_loss_no_MP = pickle.load(file)
    te_loss_no_MP = pickle.load(file)
```

### Plotting Evaluation Accuracy

```python
import matplotlib.pyplot as plt

x = range(len(tr_acc_BL))

plt.title('Evaluation Acc')
plt.plot(x, te_acc_BL, label='All BN and MP (Baseline)')
plt.plot(x, te_acc_no_BN_MP, label='no Batchnorm and Maxpool')
plt.plot(x, te_acc_no_BN, label='no Batchnorm')
plt.plot(x, te_acc_no_MP, label='no Maxpool')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```

### Plotting Evaluation Loss

```python
x = range(len(tr_acc_BL))

plt.title('Evaluation Loss')
plt.plot(x, te_loss_BL, label='All BN and MP (Baseline)')
plt.plot(x, te_loss_no_BN_MP, label='no Batchnorm and Maxpool')
plt.plot(x, te_loss_no_BN, label='no Batchnorm')
plt.plot(x, te_loss_no_MP, label='no Maxpool')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```
