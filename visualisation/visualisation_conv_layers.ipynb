## Initialising Test accuracies and loss from Pickle files

```python
import pickle
import os

file_name_BL = 'baseline_CNN.pkl' 
file_name_conv1 = 'baseline_CNN_conv_1layer.pkl' 
file_name_conv3 = 'baseline_CNN_conv_3layer.pkl' 
file_name_conv4 = 'baseline_CNN_conv_4layer.pkl' 


with open(os.getcwd() + '/../results/' + file_name_BL, 'rb') as file:
    tr_acc_BL = pickle.load(file)
    te_acc_BL = pickle.load(file)
    tr_loss_BL = pickle.load(file)
    te_loss_BL = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_conv1, 'rb') as file:
    tr_acc_C1 = pickle.load(file)
    te_acc_C1 = pickle.load(file)
    tr_loss_C1 = pickle.load(file)
    te_loss_C1 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_conv3, 'rb') as file:
    tr_acc_C3 = pickle.load(file)
    te_acc_C3 = pickle.load(file)
    tr_loss_C3 = pickle.load(file)
    te_loss_C3 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_conv4, 'rb') as file:
    tr_acc_C4 = pickle.load(file)
    te_acc_C4 = pickle.load(file)
    tr_loss_C4 = pickle.load(file)
    te_loss_C4 = pickle.load(file)
```

### Plotting Evaluation Accuracy

```python
import matplotlib.pyplot as plt

x = range(len(tr_acc_BL))

plt.title('Evaluation Acc')
plt.plot(x, te_acc_BL, label='Conv 2 (Baseline)')
plt.plot(x, te_acc_C1, label='Conv 1')
plt.plot(x, te_acc_C3, label='Conv 3')
plt.plot(x, te_acc_C4, label='Conv 4')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

### Plotting Evaluation Loss

```python
x = range(len(tr_acc_BL))

plt.title('Evaluation Loss')
plt.plot(x, te_loss_BL, label='Conv 2 (Baseline)')
plt.plot(x, te_loss_C1, label='Conv 1')
plt.plot(x, te_loss_C3, label='Conv 3')
plt.plot(x, te_loss_C4, label='Conv 4')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
