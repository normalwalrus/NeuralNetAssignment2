```python
import pickle
import os

file_name_BL = 'baseline_CNN.pkl' 
file_name_FC1 = 'baseline_CNN_FC_1.pkl' 
file_name_FC2 = 'baseline_CNN_FC_2.pkl' 
file_name_FC4 = 'baseline_CNN_FC_4.pkl' 
file_name_FC5 = 'baseline_CNN_FC_5.pkl' 


with open(os.getcwd() + '/../results/' + file_name_BL, 'rb') as file:
    tr_acc_BL = pickle.load(file)
    te_acc_BL = pickle.load(file)
    tr_loss_BL = pickle.load(file)
    te_loss_BL = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_FC1, 'rb') as file:
    tr_acc_FC1 = pickle.load(file)
    te_acc_FC1 = pickle.load(file)
    tr_loss_FC1 = pickle.load(file)
    te_loss_FC1 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_FC2, 'rb') as file:
    tr_acc_FC2 = pickle.load(file)
    te_acc_FC2 = pickle.load(file)
    tr_loss_FC2 = pickle.load(file)
    te_loss_FC2 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_FC4, 'rb') as file:
    tr_acc_FC4 = pickle.load(file)
    te_acc_FC4 = pickle.load(file)
    tr_loss_FC4 = pickle.load(file)
    te_loss_FC4 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_FC5, 'rb') as file:
    tr_acc_FC5 = pickle.load(file)
    te_acc_FC5 = pickle.load(file)
    tr_loss_FC5 = pickle.load(file)
    te_loss_FC5 = pickle.load(file)
```

### Plotting Evaluation Accuracy

```python
import matplotlib.pyplot as plt

x = range(len(tr_acc_BL))

plt.title('Evaluation Acc')
plt.plot(x, te_acc_BL, label='FC 2 (Baseline)')
plt.plot(x, te_acc_FC1, label='FC 1')
plt.plot(x, te_acc_FC2, label='FC 2')
plt.plot(x, te_acc_FC4, label='FC 4')
plt.plot(x, te_acc_FC5, label='FC 5')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

### Plotting Evaluation Loss

```python
x = range(len(tr_acc_BL))

plt.title('Evaluation Loss')
plt.plot(x, te_loss_BL, label='FC 2 (Baseline)')
plt.plot(x, te_loss_FC1, label='FC 1')
plt.plot(x, te_loss_FC2, label='FC 2')
plt.plot(x, te_loss_FC4, label='FC 4')
plt.plot(x, te_loss_FC5, label='FC 5')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
