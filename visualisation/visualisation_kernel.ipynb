## Initialising Test accuracies and loss from Pickle files

```python
import pickle
import os

file_name_BL = 'baseline_CNN.pkl' 
file_name_K1 = 'baseline_CNN_kernel_1.pkl' 
file_name_K5 = 'baseline_CNN_kernel_5.pkl' 
file_name_K7 = 'baseline_CNN_kernel_7.pkl' 

with open(os.getcwd() + '/../results/' + file_name_BL, 'rb') as file:
    tr_acc_BL = pickle.load(file)
    te_acc_BL = pickle.load(file)
    tr_loss_BL = pickle.load(file)
    te_loss_BL = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_K1, 'rb') as file:
    tr_acc_K1 = pickle.load(file)
    te_acc_K1 = pickle.load(file)
    tr_loss_K1 = pickle.load(file)
    te_loss_K1 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_K5, 'rb') as file:
    tr_acc_K5 = pickle.load(file)
    te_acc_K5 = pickle.load(file)
    tr_loss_K5 = pickle.load(file)
    te_loss_K5 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_K7, 'rb') as file:
    tr_acc_K7 = pickle.load(file)
    te_acc_K7 = pickle.load(file)
    tr_loss_K7 = pickle.load(file)
    te_loss_K7 = pickle.load(file)
```

### Plotting Evaluation Accuracy

```python
import matplotlib.pyplot as plt

x = range(len(tr_acc_BL))

plt.plot(x, te_acc_BL, label='K3 (Baseline)')
plt.plot(x, te_acc_K1, label='K1')
plt.plot(x, te_acc_K5, label='K5')
plt.plot(x, te_acc_K7, label='K7')

plt.title('Evaluation Acc')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

### Plotting Evaluation Loss

```python
x = range(len(tr_acc_BL))

plt.plot(x, te_loss_BL, label='K3 (Baseline)')
plt.plot(x, te_loss_K1, label='K1')
plt.plot(x, te_loss_K5, label='K5')
plt.plot(x, te_loss_K7, label='K7')

plt.title('Evaluation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
