```python
import pickle
import os

file_name_BL = 'baseline_CNN.pkl' 
file_name_RN18 = 'pretrained_resnet18_nofreeze.pkl' 
file_name_RN50 = 'pretrained_resnet50_nofreeze.pkl' 
file_name_GN = 'pretrained_googlenet_nofreeze.pkl'  
file_name_Alex = 'pretrained_alexnet_nofreeze.pkl' 

with open(os.getcwd() + '/../results/' + file_name_BL, 'rb') as file:
    tr_acc_BL = pickle.load(file)
    te_acc_BL = pickle.load(file)
    tr_loss_BL = pickle.load(file)
    te_loss_BL = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_RN18, 'rb') as file:
    tr_acc_RN18 = pickle.load(file)
    te_acc_RN18 = pickle.load(file)
    tr_loss_RN18 = pickle.load(file)
    te_loss_RN18 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_GN, 'rb') as file:
    tr_acc_GN = pickle.load(file)
    te_acc_GN = pickle.load(file)
    tr_loss_GN = pickle.load(file)
    te_loss_GN = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_Alex, 'rb') as file:
    tr_acc_Alex = pickle.load(file)
    te_acc_Alex = pickle.load(file)
    tr_loss_Alex = pickle.load(file)
    te_loss_Alex = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_RN50, 'rb') as file:
    tr_acc_RN50 = pickle.load(file)
    te_acc_RN50 = pickle.load(file)
    tr_loss_RN50 = pickle.load(file)
    te_loss_RN50 = pickle.load(file)
```

### Plotting Evaluation Accuracy

```python
import matplotlib.pyplot as plt

x = range(len(tr_acc_BL))

plt.plot(range(len(te_acc_BL)), te_acc_BL, label='Baseline')
plt.plot(range(len(te_acc_RN18)), te_acc_RN18, label='ResNet18')
plt.plot(range(len(te_acc_GN)), te_acc_GN, label='GoogleNet')
plt.plot(range(len(te_acc_Alex)), te_acc_Alex, label='AlexNet')
plt.plot(range(len(te_acc_RN50)), te_acc_RN50, label='ResNet50')

plt.title('Evaluation Acc')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

### Plotting Evaluation Loss

```python
x = range(len(tr_acc_BL))

plt.plot(range(len(te_loss_BL)), te_loss_BL, label='Baseline')
plt.plot(range(len(te_loss_RN18)), te_loss_RN18, label='ResNet18')
plt.plot(range(len(te_loss_GN)), te_loss_GN, label='GoogleNet')
plt.plot(range(len(te_loss_Alex)), te_loss_Alex, label='AlexNet')
plt.plot(range(len(te_loss_RN50)), te_loss_RN50, label='ResNet50')

plt.title('Evaluation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
