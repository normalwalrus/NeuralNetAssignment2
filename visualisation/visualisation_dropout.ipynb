```python
import pickle
import os

file_name_BL = 'baseline_CNN.pkl' 
file_name_DO1 = 'baseline_CNN_dropout0.5.pkl' 
file_name_DO2 = 'baseline_CNN_dropout0.75.pkl' 
file_name_DO3 = 'baseline_CNN_dropout0.pkl' 


with open(os.getcwd() + '/../results/' + file_name_BL, 'rb') as file:
    tr_acc_BL = pickle.load(file)
    te_acc_BL = pickle.load(file)
    tr_loss_BL = pickle.load(file)
    te_loss_BL = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_DO1, 'rb') as file:
    tr_acc_DO1 = pickle.load(file)
    te_acc_DO1 = pickle.load(file)
    tr_loss_DO1 = pickle.load(file)
    te_loss_DO1 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_DO2, 'rb') as file:
    tr_acc_DO2 = pickle.load(file)
    te_acc_DO2 = pickle.load(file)
    tr_loss_DO2 = pickle.load(file)
    te_loss_DO2 = pickle.load(file)

with open(os.getcwd() + '/../results/' + file_name_DO3, 'rb') as file:
    tr_acc_DO3 = pickle.load(file)
    te_acc_DO3 = pickle.load(file)
    tr_loss_DO3 = pickle.load(file)
    te_loss_DO3 = pickle.load(file)
```

### Plotting Evaluation Accuracy

```python
import matplotlib.pyplot as plt

x = range(len(tr_acc_BL))

plt.title('Evaluation Acc')
plt.plot(x, te_acc_BL, label='Dropout 0.25(Baseline)')
plt.plot(x, te_acc_DO1, label='Dropout 0.5')
plt.plot(x, te_acc_DO2, label='Dropout 0.75')
plt.plot(x, te_acc_DO3, label='Dropout 0')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```

### Plotting Evaluation Loss

```python
x = range(len(tr_acc_BL))

plt.title('Evaluation Loss')
plt.plot(x, te_loss_BL, label='Dropout 0.25(Baseline)')
plt.plot(x, te_loss_DO1, label='Dropout 0.5')
plt.plot(x, te_loss_DO2, label='Dropout 0.75')
plt.plot(x, te_loss_DO3, label='Dropout 0')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```
